```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelBinarizer
import tensorflow as tf
from tensorflow import keras

print(tf.__version__)
```

```python
train_df = pd.read_csv('train.csv')
```

```python
train_df.info()
```

```python
# eliminar colunas que nao tem dados suficientes
train_df = train_df.drop('Alley', axis=1)
train_df = train_df.drop('FireplaceQu', axis=1)
train_df = train_df.drop('PoolQC', axis=1)
train_df = train_df.drop('Fence', axis=1)
train_df = train_df.drop('MiscFeature', axis=1)
```

```python
train_df.head(10)
```

```python
# preencher valores nulos (NaN) com a media dos valores da respectiva coluna
train_df = train_df.fillna(train_df.mean())
train_df.head(10)
```

```python
train_df.info()
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputed_DF = train_df
for col in train_df.columns:
    if (train_df[col].isna().any()):
        imputed_column = imp.fit_transform(np.array(train_df[col]).reshape(-1, 1))
        #Fill in Series on DataFrame
        imputed_DF[col] = imputed_column
        
train_df = imputed_DF
imputed_DF.info()
```

```python
categorical_feature_mask = train_df.dtypes==object
categorical_cols = train_df.columns[categorical_feature_mask].tolist()
categorical_cols
```

```python
concat_df = train_df
lb_style = LabelBinarizer()
for col in categorical_cols:
    lb_results = lb_style.fit_transform(train_df[col])
    if (lb_results.shape[1] == 1):
        concat_df[col] = lb_results
    else:
        lb_df = pd.DataFrame(lb_results, columns=lb_style.classes_)
        for cls in lb_style.classes_:
            concat_df[cls] = lb_df[cls]
        concat_df = concat_df.drop(col, axis=1)
train_df = concat_df    
```
